#### Check the distribution of variables

```python
def draw_histograms(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax,facecolor='purple')
        ax.set_title(feature+" Distribution",color='DarkRed')
        ax.set_yscale('log')
    fig.tight_layout()  
    plt.show()
draw_histograms(creditcard,creditcard.columns,8,4)
```

#### Visualizing the correlation matrix for better readability and understanding.

```python
plt.figure(figsize=(40, 15))
heatmap=sns.heatmap(creditcard.corr(),vmin=-1,vmax=1,annot=True,cmap='PiYG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);

```
